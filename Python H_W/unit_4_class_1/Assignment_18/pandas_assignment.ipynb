# Task 5: Grouping and Filtering Data

Grouped Village

```python
df.groupby("Village")["Power Level"].mean()
```

```python
grouped = df.groupby("Village")
print(grouped)
```

```python
max_level = grouped["Power Level"].max()
max_level
```

```python
min_level = grouped["Power Level"].min()
min_level
```

Filtered Power

```python
df[df["Power Level"] > 90]
```

```python
df[df["Power Level"] < 95]
```

Filtered Name

```python
df[df["Name"].str.contains("Hatake", case=False)]
```

```python
df[df["Clan"].str.contains("Uzumaki", case=False)]
```
