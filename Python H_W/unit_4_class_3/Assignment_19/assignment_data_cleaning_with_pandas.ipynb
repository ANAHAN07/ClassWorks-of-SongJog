# 1. Handling Missing Values:

# 2. Fixing Inconsistent Data

* Standardize gender values (e.g., 'M', 'Male', 'male' should be uniform).

```python
df['Gender'] = df['Gender'].str.strip().str.lower()
df['Gender'] = df['Gender'].replace({'male': 'M', 'm': 'M', 'female': 'F', 'f': 'F'})

df
```

* Convert Join_Date into a standard date format.

```python
df['Join_Date'] = pd.to_datetime(df['Join_Date'], errors='coerce')
df
```

# 3. Handling Duplicates

* Identify and remove duplicate records if necessary.

```python
print("Duplicates before removal:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
print("Duplicates after removal:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

# 4. Detecting and Treating Outliers:

* Check for outliers in the Salary column and handle them appropriately.

 A Function To Remove Outliers Using IQR

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df


df
```

```python
df = remove_outliers(df, 'Salary')
df
```

# 5. Data Type Conversion

* Ensure all columns have appropriate data types.

```python
df['Salary'] = df['Salary'].astype(int)

df
```
